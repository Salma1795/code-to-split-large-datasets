# **Split into Equal-Sized Chunks**

```python
import numpy as np
import os
# Number of chunks you want to split the data into
num_chunks = 10

# Split the dataset into num_chunks parts
chunks = np.array_split(df, num_chunks)

# Specify the directory where you want to save the files
output_dir = '/content/drive/MyDrive/split try'

# Create the directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)
```

# Save the Chunks to Separate Files.
**Once you have the chunks, you can save them as separate files, such as CSV or pickle files.**

```python
# Save each chunk as a CSV file in the specified directory
for i, chunk in enumerate(chunks):
    chunk.to_csv(f'{output_dir}/salary{i}.csv', index=False)

print(f"All chunks saved in {output_dir}")
```
